```python
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# Load your cleaned CVE data
# cve_data_cleaned = pd.read_csv('path_to_your_cleaned_data.csv')
# For example:
cve_data= pd.read_csv('/kaggle/input/cve-data/cve_data_cleaned.csv')


```

```python
# Initialize the model
model = SentenceTransformer('sentence-transformers/all-mpnet-base-v2')

# Convert descriptions to embeddings
cve_data['Embeddings'] = cve_data['Description'].apply(lambda x: model.encode(x, convert_to_tensor=True))

# If needed, you can convert the embeddings into a numpy array for further processing
embeddings = np.vstack(cve_data['Embeddings'].to_numpy())
```

```python
import pandas as pd
from sentence_transformers import SentenceTransformer

# Initialize the model
model = SentenceTransformer('sentence-transformers/all-mpnet-base-v2')

# Function to process in batches
def batch_encode(data, batch_size=32):
    embeddings = []
    for i in range(0, len(data), batch_size):
        batch = data[i:i + batch_size]
        # Convert to embeddings and move to CPU
        batch_embeddings = model.encode(batch.tolist(), convert_to_tensor=True)
        # Convert to NumPy array and add to the list
        embeddings.extend(batch_embeddings.cpu().numpy())
    return embeddings

# Convert descriptions to embeddings in batches
cve_data['Embeddings'] = batch_encode(cve_data['Description'])
```

```python
import pandas as pd

# Assuming 'Embeddings' contains your embeddings and 'Description' is also relevant
# You can include any other columns you want to save
embeddings_df = pd.DataFrame(cve_data['Embeddings'].tolist())  # Convert embeddings to DataFrame
embeddings_df['Description'] = cve_data['Description'].values  # Add Description column

# Save to CSV
embeddings_df.to_csv('cve_embeddings.csv', index=False)
```
